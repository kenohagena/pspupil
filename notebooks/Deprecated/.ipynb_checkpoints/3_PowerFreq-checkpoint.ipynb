```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyedfread import edf
import pickle
from glob import glob
from os.path import join
%matplotlib inline
import seaborn as sns
from scipy import signal
from scipy import interpolate
#from PSPupil import pupil_preprocess as psp
#from PSPupil import PupilPreprocess as PS
from PSPupil import Preprocessing as PS
from PSPupil import dataset as dt

from scipy import signal
import mne
from scipy.stats import ttest_ind, linregress
from scipy.optimize import curve_fit
import svgutils.transform as sg


# plotting parameters
control_color='teal'
baseline_color='goldenrod'
follow_color='indianred'

def smaller_than(value):
    logs=np.append(np.logspace(-15, 1, 17), .05)
    sorteds=np.sort(np.append(logs, value))
    index=np.where(sorteds==value)[0]
    return sorteds[index+1][0]

def func_powerlaw(x, m, c, c0):
    return c0 + x**m * c

def permutation_cluster_test(x1, x2):
    f, clusters, pvalues, h0=mne.stats.permutation_cluster_test([x1,x2],
                                                                     threshold=dict(start=0, step=0.2))
    return pvalues
```

```python
lod = []
for sub, dic in dt.runs.items():
    for session, lis in dic.items():
        if lis is None:
            print(None)
        else:
            for i, run in enumerate(lis):
                #print(sub, session, run)
                data= pd.read_hdf(join(edition,'patient', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}.hdf'.format(sub, session, run)), key='Pupil')
                x = data.biz.fillna(method='bfill').values
                #plt.plot(x)
                #plt.show()
                f, p = signal.welch(x, fs=freq, nperseg=segment*freq, noverlap=overlap*freq, detrend='constant')
                fpower = dict(zip(f, p))
                fpower['subject'] = sub
                fpower['session'] = session
                fpower['run'] = i
                lod.append(fpower)
p_fpower = pd.DataFrame(lod)
p_fpower.to_hdf(join(edition, 'power_freqs_{0}_{1}.hdf'.format(segment, overlap)), key='data')
```

## Calculate AUC

```python
y = func_powerlaw(auc_x, *popt)
```

```python
control_auc=[]
for c in control.index:
    y1 = control.loc[c, auc_x]
    d = y1.astype(float) - y
    auc = np.trapz(d, x=auc_x)
    control_auc.append({'subject': c,
                       'group': 'control',
                       'auc':auc})
    #print(auc)
control_auc = pd.DataFrame(control_auc)
```

```python
patient_auc=[]
for c in baseline.index:
    #print(c)

    y1 = baseline.loc[c, auc_x]

    d = y1.astype(float) - y
    patient_auc.append({'subject': c, 
                        'group':'baseline',
                        'auc':np.trapz(d, x=auc_x)})
patient_auc = pd.DataFrame(patient_auc)
```

```python
follow_auc=[]
for c in followup.index:
    #print(c)

    y1 = followup.loc[c, auc_x]

    d = y1.astype(float) - y
    follow_auc.append({'subject': c, 
                        'group':'follow',
                        'auc':np.trapz(d, x=auc_x)})
follow_auc = pd.DataFrame(follow_auc)
```

### Find hippus peak

### Coefficient of variation (CoV)

https://www.sciencedirect.com/science/article/pii/S016787601831105X

```python
CoV = []
```

```python
for sub, lis in dt.c_runs.items():
    session = 'Baseline'
    if lis is None:
        print(None)
    else:
        for run in lis:
            data= pd.read_hdf(join(edition, 'control', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}.hdf'.format(sub, session, run)), key='Pupil')
            di = {'group': 'control',
                'session': session,

                'subject': sub,
                 'run': run,
                 'mean': data.d_intp.mean(),
                 'std': data.d_intp.std()}
            CoV.append(di)
            
for sub, dic in dt.runs.items():
    for session in ['Baseline', 'Followup']:
        for session, lis in dic.items():

            if lis is None:
                #print(None)
                continue
            else:
                for i, run in enumerate(lis):

                    #print(sub, session, run)
                    data= pd.read_hdf(join(edition,'patient', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}.hdf'.format(sub, session, run)), key='Pupil')
                    di = {'group': 'psp',
                          'session': session,
                    'subject': sub,
                     'run': run,
                     'mean': data.d_intp.mean(),
                     'std': data.d_intp.std()}
                CoV.append(di)
            
for sub, lis in dt.ips_runs.items():
    if sub == '12':
        continue
    else:
        session = 'STIMOFF'
        for run in lis:
            data= pd.read_hdf(join(edition, 'ips', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}.hdf'.format(sub, session, run)), key='Pupil')
            di = {'group': 'ips',
                  'session': session,
                'subject': sub,
                 'run': run,
                 'mean': data.d_intp.mean(),
                 'std': data.d_intp.std()}
            CoV.append(di)
```
