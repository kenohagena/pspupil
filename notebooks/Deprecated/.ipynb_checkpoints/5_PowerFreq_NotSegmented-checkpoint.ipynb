```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyedfread import edf
import pickle
from glob import glob
from os.path import join
%matplotlib inline
import seaborn as sns
from scipy import signal
from scipy import interpolate
#from PSPupil import pupil_preprocess as psp
#from PSPupil import PupilPreprocess as PS
from PSPupil import Preprocessing as PS
from PSPupil import dataset as dt

from scipy import signal
import mne
from scipy.stats import ttest_ind, linregress, f_oneway#, tukey_hsd
from scipy.optimize import curve_fit
import svgutils.transform as sg
from decim.adjuvant import slurm_submit as slu
import matplotlib.colors
import itertools
import scipy



def smaller_than(value):
    logs=np.append(np.logspace(-15, 1, 17), .05)
    sorteds=np.sort(np.append(logs, value))
    index=np.where(sorteds==value)[0]
    return sorteds[index+1][0]

def func_powerlaw(x, m, c, c0):
    return c0 + x**m * c

def permutation_cluster_test(x1, x2):
    f, clusters, pvalues, h0=mne.stats.permutation_cluster_test([x1,x2],
                                                                     threshold=dict(start=0, step=0.2))
    return pvalues

def rgb_to_hex(r, g, b):
    return '#{:02x}{:02x}{:02x}'.format(r, g, b)
```

```python
# parameters for FFT

freq = 60
segment = 60
overlap = segment/2
```

```python
# calculate CoV weiter unten
CoV = []
```

```python
# Calculate FFT for PSP patients 

d = []
for subject, dictionary in dt.runs.items():
    for session, runs in dictionary.items():
        if runs != None:
            for run in runs:
                if float(subject) < 27:
                    files = glob(join(edition, 'PSP', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}*.hdf'.format(subject, session, run)))
                    for file in files:
                        data = pd.read_hdf(file, key='Pupil')
                        
                        x = data.biz.fillna(method='bfill').values
                        f, p = signal.welch(x, fs=freq, nperseg=segment*freq, noverlap=overlap*freq, detrend='constant')
                        fpower = dict(zip(f, p))
                        fpower['subject'] = subject
                        fpower['session'] = session
                        fpower['run'] = run
                        d.append(fpower)
                        
                        di = {'group': 'PSP',
                          'session': session,
                        'subject': subject,
                         'run': run,
                         'mean': data.d_intp.mean(),
                         'std': data.d_intp.std()}
                        CoV.append(di)

                        
psp_fpower = pd.DataFrame(d)
psp_fpower.to_hdf(join(edition, 'PowerFrequencies', 'power_freqs_{0}_{1}.hdf'.format(segment, overlap)), key='data')
```

```python
# Calculate FFT for helathy controls 

d = []
for subject, runs in dt.c_runs.items():
    for run in runs:
        if subject[0] != 'H':
            group = 'Controls'
            session = 'Baseline'
        
            files = glob(join(edition, 'Controls', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}*.hdf'.format(subject, session, run)))
            for file in files:
                data = pd.read_hdf(file, key='Pupil')
                x = data.biz.fillna(method='bfill').values
                f, p = signal.welch(x, fs=freq, nperseg=segment*freq, noverlap=overlap*freq, detrend='constant')
                fpower = dict(zip(f, p))
                fpower['subject'] = subject
                fpower['session'] = session
                fpower['run'] = run
                d.append(fpower)
                
                di = {'group': 'Controls',
                  'session': session,
                'subject': subject,
                 'run': run,
                 'mean': data.d_intp.mean(),
                 'std': data.d_intp.std()}
                CoV.append(di)
                
ctrl_fpower = pd.DataFrame(d)
ctrl_fpower.to_hdf(join(edition, 'PowerFrequencies', 'power_freqs_{0}_{1}_ctrl.hdf'.format(segment, overlap)), key='data')
```

```python
# Calculate FFT for PD patients 

d = []
for subject, runs in dt.ips_runs.items():
    for run in runs:
        if subject[0] != 'H':
            group = 'PD'
            session = 'Baseline'
        
            files = glob(join(edition, 'PD', 'Pupil_Preprocessed_SUB-{0}_*_{1}*.hdf'.format(subject, run)))
            for file in files:
                data = pd.read_hdf(file, key='Pupil')
                x = data.biz.fillna(method='bfill').values
                f, p = signal.welch(x, fs=freq, nperseg=segment*freq, noverlap=overlap*freq, detrend='constant')
                fpower = dict(zip(f, p))
                fpower['subject'] = subject
                fpower['session'] = session
                fpower['run'] = run
                d.append(fpower)
                
                di = {'group': 'PD',
                  'session': session,
                'subject': subject,
                 'run': run,
                 'mean': data.d_intp.mean(),
                 'std': data.d_intp.std()}
                CoV.append(di)

pd_fpower = pd.DataFrame(d)
pd_fpower.to_hdf(join(edition, 'PowerFrequencies', 'power_freqs_{0}_{1}_pd.hdf'.format(segment, overlap)), key='data')
```

```python
# above calculations are saved, directly import to skip new calculations

ctrl_fpower = pd.read_hdf(join(edition, 'PowerFrequencies', 'power_freqs_{0}_{1}_ctrl.hdf'.format(segment, overlap)), key='data')
psp_fpower= pd.read_hdf(join(edition, 'PowerFrequencies','power_freqs_{0}_{1}.hdf'.format(segment, overlap)), key='data')
pd_fpower= pd.read_hdf(join(edition, 'PowerFrequencies','power_freqs_{0}_{1}_pd.hdf'.format(segment, overlap)), key='data')
```

```python
# avergae per subject (and session = baseline or followup)

psp_fpower_avg = psp_fpower.groupby(['subject', 'session']).mean().reset_index()
ctrl_fpower_avg = ctrl_fpower.groupby(['subject', 'session']).mean().reset_index()
pd_fpower_avg = pd_fpower.groupby(['subject', 'session']).mean().reset_index()

# split baseline and followup for psp and set index

psp_baseline = psp_fpower_avg.loc[psp_fpower_avg.session == 'Baseline'].set_index('subject')
psp_followup = psp_fpower_avg.loc[psp_fpower_avg.session == 'Followup'].set_index('subject')
ctrl_fpower_avg = ctrl_fpower_avg.set_index('subject')
pd_fpower_avg = pd_fpower_avg.set_index('subject')
```

```python
#data for permutation testint
d = psp_baseline.drop(['session'], axis=1).unstack().reset_index()
d.level_0 = d.level_0.astype(float)
base = d.loc[(d.level_0 >=1/32)&(d.level_0 <=1)]
base = base.pivot(columns = 'level_0', index='subject')

d = ctrl_fpower_avg.drop(['session'], axis=1).unstack().reset_index()
d.level_0 = d.level_0.astype(float)
cont = d.loc[(d.level_0 >=1/32)&(d.level_0 <=1)]
cont = cont.pivot(columns = 'level_0', index='subject')

d = psp_followup.drop(['session'], axis=1).unstack().reset_index()
d.level_0 = d.level_0.astype(float)
fol = d.loc[(d.level_0 >=1/32)&(d.level_0 <=1)]
fol = fol.pivot(columns = 'level_0', index='subject')

d = pd_fpower_avg.drop(['session'], axis=1).unstack().reset_index()
d.level_0 = d.level_0.astype(float)
pdg = d.loc[(d.level_0 >=1/32)&(d.level_0 <=1)]
pdg = pdg.pivot(columns = 'level_0', index='subject')

permut_x = pdg.droplevel(0, axis=1).columns

pvals_psp_c = permutation_cluster_test(base.values, 
                         cont.values)
print(pvals_psp_c<.05)


pvals_psp_pd = permutation_cluster_test(base.values, 
                         pdg.values)
print(pvals_psp_pd<.05)


pvals_pd_cont = permutation_cluster_test(cont.values, 
                         pdg.values)
print(pvals_pd_cont<.05)
```

```python
#fit power law to both control and patient data as reference to later calculate AUC etc.

a = ctrl_fpower_avg.groupby(['subject', 'session']).mean().stack().reset_index().rename(columns={'level_2': 'freq',
                                                                                       0: 'power'})
b = psp_baseline.groupby(['subject', 'session']).mean().stack().reset_index().rename(columns={'level_2': 'freq',
                                                                                       0: 'power'})
c = pd_fpower_avg.groupby(['subject', 'session']).mean().stack().reset_index().rename(columns={'level_2': 'freq',
                                                                                       0: 'power'})

combined=pd.concat([a, b, c], axis=0)
combined.freq = combined.freq.astype(float)

combined = combined.loc[(combined.freq > 1/32)&(combined.freq < 1)]

X=combined.freq.values
y=combined.power.values
popt, pcov = curve_fit(func_powerlaw, X, y, p0 = np.asarray([-1,10**5,0]), maxfev=800)

X=combined.freq.unique()
plt.plot(X, func_powerlaw(X, *popt), color='black', linestyle='--')

plt.plot(ctrl_fpower_avg.groupby(['subject', 'session']).mean().reset_index().groupby('subject').mean().mean().index,
         ctrl_fpower_avg.groupby(['subject', 'session']).mean().reset_index().groupby('subject').mean().mean().values,
         color='green', alpha=.5)

plt.plot(psp_fpower_avg.groupby(['subject', 'session']).mean().reset_index().groupby('subject').mean().mean().index,
         psp_fpower_avg.groupby(['subject', 'session']).mean().reset_index().groupby('subject').mean().mean().values,
         color='red', alpha=.5)

plt.plot(pd_fpower_avg.groupby(['subject', 'session']).mean().reset_index().groupby('subject').mean().mean().index,
         pd_fpower_avg.groupby(['subject', 'session']).mean().reset_index().groupby('subject').mean().mean().values,
         color='blue', alpha=.5)

plt.plot(combined.groupby('freq').mean().reset_index().freq, combined.groupby('freq').mean().reset_index().power, color='black')


# Slope / AUC function
x = ctrl_fpower_avg.columns.drop(['session'])
auc_x = x[(x<=1/2)&(x>=1/16)].astype(float)

x[np.where(x>1/12)[0][0]:np.where(x>1/3)[0][0]].astype(float) #for auc reference


plt.xlim(0, 1)
plt.ylim(0, 2)
```

```python
# calculate Hippus for PSP baseline

f, ax = plt.subplots(5, 4, figsize=(15, 12))
f.tight_layout()
data=psp_baseline.drop(['session'], axis=1)

i=0
for alist in ax:
    for a in alist:
        try:
            sub = data.index[i]
            d = data.loc[sub]
            d = d.loc[(d.index >= 1/32)&(d.index <= 1)]
            X = d.index.astype(float)
            y = d.values

            popt, pcov = curve_fit(func_powerlaw, X, y, p0 = np.asarray([-1,10**5,0]), maxfev=2000)

            a.plot(X, func_powerlaw(X, *popt), '--', color='green')

            d1 = d.loc[(d.index > 1/10)&(d.index < 1)] # q&d way to exclude maxima in the descending power law spectrum
            X1 = d1.index.astype(float)
            y1 = d1.values
            hippus_freq = X1[np.argmax(y1-func_powerlaw(X1, *popt))]
            power_offset = np.max(y1-func_powerlaw(X1, *popt))
            
            di = {'subject':sub,
                          'group':'baseline',
                          'frequency':hippus_freq,
                          'offset':power_offset}
            if hippus_freq > hip_limit:
                di['frequency'] = np.nan
            hippus.append(di)
            a.set(title='Subject {}'.format(sub))


            a.plot(d.index, d.values, color='grey')
            a.axvline(di['frequency'], color='black')
            '''
            if power_offset < 50*np.mean(y1-func_powerlaw(X1, *popt)):
                print(sub)
            '''
            i=i+1
        except IndexError:
            continue
```

```python
# table with gait characteristics vPSP vs PSP-RS
cols = ['Velocity', 'Cadence', 'StepLengthmean',
       'CycleTimemean', 'SingleSupportmean', 'DoubleSupportmean', 'Swingmean', 'StepLengthSymmetryRatio']

mean = gcp.loc[~gcp.Velocity.isnull()].groupby(['condition', 'group']).mean().loc[:, cols].T
std = gcp.loc[~gcp.Velocity.isnull()].groupby(['condition', 'group']).std().loc[:, cols].T

for condition in ['fast', 'normal', 'turn']:
    for param in mean.index:
        samples = [gcp.loc[(gcp.condition == condition) & (gcp.group == 'PSP-RS'), param].values,
                    gcp.loc[(gcp.condition == condition) & (gcp.group == 'vPSP'), param].values,
                    gcp.loc[(gcp.condition == condition) & (gcp.group == 'PD'), param].values,
                    gcp.loc[(gcp.condition == condition) & (gcp.group == 'HC'), param].values]
        anova = f_oneway(samples[0],samples[1],samples[2],samples[3])
        if anova[1]<0.05:
            print(condition, param)
        mean.loc[param, (condition, 'F')] = anova[0]
        mean.loc[param, (condition, 'p_value')] = anova[1]
        mean.loc[param, (condition, 't_test_RS_vPSP')] = ttest_ind(gcp.loc[(gcp.condition == condition) & (gcp.group == 'PSP-RS'), param].values,
                    gcp.loc[(gcp.condition == condition) & (gcp.group == 'vPSP'), param].values)[1]
```
