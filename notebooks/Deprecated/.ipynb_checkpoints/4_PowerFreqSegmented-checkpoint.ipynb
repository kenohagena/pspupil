```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyedfread import edf
import pickle
from glob import glob
from os.path import join
%matplotlib inline
import seaborn as sns
from scipy import signal
from scipy import interpolate
#from PSPupil import pupil_preprocess as psp
#from PSPupil import PupilPreprocess as PS
from PSPupil import Preprocessing as PS
from PSPupil import dataset as dt

from scipy import signal
import mne
from scipy.stats import ttest_ind, linregress
from scipy.optimize import curve_fit
import svgutils.transform as sg
from decim.adjuvant import slurm_submit as slu
import matplotlib.colors

#Plotting parameters
palette = sns.color_palette("Paired")

baseline_color = palette[4]
followup_color = palette[5]
ips_color = palette[0]
control_color = palette[2]

plt.rcParams.update({'font.size': 16})
psize=12
def smaller_than(value):
    logs=np.append(np.logspace(-15, 1, 17), .05)
    sorteds=np.sort(np.append(logs, value))
    index=np.where(sorteds==value)[0]
    return sorteds[index+1][0]

def func_powerlaw(x, m, c, c0):
    return c0 + x**m * c

def permutation_cluster_test(x1, x2):
    f, clusters, pvalues, h0=mne.stats.permutation_cluster_test([x1,x2],
                                                                     threshold=dict(start=0, step=0.2))
    return pvalues
```

## Description

- analysis using Preprocessed Data applying Approach with segmentation (do not include invalid segments >350 ms and avoid long interpolation), therefore multiple segments per run
- do FFT on each segment, than average weighted by segment length
- 1. weighted average per run, 2. weighted average per session

```python
edition = '/Volumes/psp_data/Preprocessing_segmented_2023-12-27' #edition with segmented runs
non_edition = '/Volumes/psp_data/Pupil_Preprocessed_2023-12' #edition with non-segmentation to compare

slu.mkdir_p(join(edition, 'PowerFrequencies'))
```

```python
freq = 60
segment = 24
overlap = segment/2
```

```python
# do FFT on segmented data

d = []
for subject, dictionary in dt.runs.items():
    for session, runs in dictionary.items():
        if runs != None:
            for run in runs:
                if float(subject) < 27:
                    files = glob(join(edition, 'PSP', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}*.hdf'.format(subject, session, run)))
                    for file in files:
                        f = '{0}_{1}'.format(session, run)
                        segment_no= file[file.find(f)+ len(f)+1:-4]
                        data = pd.read_hdf(file, key='Pupil')
                        length = data.index[-1]- data.index[0]
                        length = length.total_seconds()
                        if length > 25: # somehow 24s/segments are too short
                            #print(subject, session, run, segment_no, length)
                            x = data.biz.fillna(method='bfill').values
                            f, p = signal.welch(x, fs=freq, nperseg=segment*freq, noverlap=overlap*freq, detrend='constant')
                            fpower = dict(zip(f, p))
                            fpower['subject'] = subject
                            fpower['session'] = session
                            fpower['run'] = run
                            fpower['segment'] = segment_no
                            fpower['length'] = length
                            d.append(fpower)
                        
psp_fpower = pd.DataFrame(d)
psp_fpower.to_hdf(join(edition, 'PowerFrequencies', 'power_freqs_{0}_{1}.hdf'.format(segment, overlap)), key='data')


d = []
for subject, runs in dt.c_runs.items():
    for run in runs:
        if subject[0] != 'H':
            sub = subject
            group = 'Controls'
            session = 'Baseline'
        
            files = glob(join(edition, 'Controls', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}*.hdf'.format(subject, session, run)))
            for file in files:
                f = '{0}_{1}'.format(session, run)
                segment_no= file[file.find(f)+ len(f)+1:-4]
                data = pd.read_hdf(file, key='Pupil')
                length = data.index[-1]- data.index[0]
                length = length.total_seconds()
                if length > 25: # somehow 24s/segments are too short
                    x = data.biz.fillna(method='bfill').values
                    f, p = signal.welch(x, fs=freq, nperseg=segment*freq, noverlap=overlap*freq, detrend='constant')
                    fpower = dict(zip(f, p))
                    fpower['subject'] = sub
                    fpower['session'] = session
                    fpower['run'] = run
                    fpower['segment'] = segment_no
                    fpower['length'] = length
                    d.append(fpower)

ctrl_fpower = pd.DataFrame(d)
ctrl_fpower.to_hdf(join(edition, 'PowerFrequencies', 'power_freqs_{0}_{1}_ctrl.hdf'.format(segment, overlap)), key='data')

d = []
for subject, runs in dt.c_runs.items():
    for run in runs:
        if subject[0] != 'H':
            sub = subject
            group = 'Controls'
            session = 'Baseline'
        
            files = glob(join(edition, 'Controls', 'Pupil_Preprocessed_SUB-{0}_{1}_{2}*.hdf'.format(subject, session, run)))
            for file in files:
                f = '{0}_{1}'.format(session, run)
                segment_no= file[file.find(f)+ len(f)+1:-4]
                data = pd.read_hdf(file, key='Pupil')
                length = data.index[-1]- data.index[0]
                length = length.total_seconds()
                if length > 25: # somehow 24s/segments are too short
                    x = data.biz.fillna(method='bfill').values
                    f, p = signal.welch(x, fs=freq, nperseg=segment*freq, noverlap=overlap*freq, detrend='constant')
                    fpower = dict(zip(f, p))
                    fpower['subject'] = sub
                    fpower['session'] = session
                    fpower['run'] = run
                    fpower['segment'] = segment_no
                    fpower['length'] = length
                    d.append(fpower)

ctrl_fpower = pd.DataFrame(d)
ctrl_fpower.to_hdf(join(edition, 'PowerFrequencies', 'power_freqs_{0}_{1}_ctrl.hdf'.format(segment, overlap)), key='data')
```

```python
# 1.1 average in psp per run across segments, weighted per segment length

di = []
for subject in psp_fpower.subject.unique():
    for session in psp_fpower.loc[psp_fpower.subject == subject].session.unique():
        for run in psp_fpower.loc[(psp_fpower.subject == subject)&(psp_fpower.session==session)].run.unique():
            d = psp_fpower.loc[(psp_fpower.subject == subject)&(psp_fpower.session==session)&(psp_fpower.run == run)].set_index(['subject', 'session', 'run', 'segment'])
            l = d.length.values
            d = d.drop('length', axis=1)
            balanced_avg = (d.values.T * l).T.mean(axis=0)/(l.sum()/2)
            d = dict(zip(d.columns, balanced_avg))
            d['subject'] = subject
            d['session'] = session
            d['run'] = run
            d['length'] = l.sum()
            di.append(d)
psp_fpower_avgrun = pd.DataFrame(di)

#1.2 average in psp per session across runs, weighted per sum-length of segments per run

di = []
for subject in psp_fpower_avgrun.subject.unique():
    for session in psp_fpower_avgrun.loc[psp_fpower_avgrun.subject == subject].session.unique():
        d = psp_fpower_avgrun.loc[(psp_fpower_avgrun.subject == subject)&(psp_fpower_avgrun.session==session)].set_index(['subject', 'session', 'run'])
        l = d.length.values
        d = d.drop('length', axis=1)
        balanced_avg = (d.values.T * l).T.mean(axis=0)/(l.sum()/2)
        d = dict(zip(d.columns, balanced_avg))
        d['subject'] = subject
        d['session'] = session
        d['length'] = l.sum()
        di.append(d)
psp_fpower_avg = pd.DataFrame(di)

# 2. the same for ctrl

di = []
for subject in ctrl_fpower.subject.unique():
    for session in ctrl_fpower.loc[ctrl_fpower.subject == subject].session.unique():
        for run in ctrl_fpower.loc[(ctrl_fpower.subject == subject)&(ctrl_fpower.session==session)].run.unique():
            d = ctrl_fpower.loc[(ctrl_fpower.subject == subject)&(ctrl_fpower.session==session)&(ctrl_fpower.run == run)].set_index(['subject', 'session', 'run', 'segment'])
            l = d.length.values
            d = d.drop('length', axis=1)
            balanced_avg = (d.values.T * l).T.mean(axis=0)/(l.sum()/2)
            d = dict(zip(d.columns, balanced_avg))
            d['subject'] = subject
            d['session'] = session
            d['run'] = run
            d['length'] = l.sum()
            di.append(d)
ctrl_fpower_avgrun = pd.DataFrame(di)


di = []
for subject in ctrl_fpower_avgrun.subject.unique():
    for session in ctrl_fpower_avgrun.loc[ctrl_fpower_avgrun.subject == subject].session.unique():
        d = ctrl_fpower_avgrun.loc[(ctrl_fpower_avgrun.subject == subject)&(ctrl_fpower_avgrun.session==session)].set_index(['subject', 'session', 'run'])
        l = d.length.values
        d = d.drop('length', axis=1)
        balanced_avg = (d.values.T * l).T.mean(axis=0)/(l.sum()/2)
        d = dict(zip(d.columns, balanced_avg))
        d['subject'] = subject
        d['session'] = session
        d['length'] = l.sum()
        di.append(d)
ctrl_fpower_avg = pd.DataFrame(di)

# 3. and for PD

di = []
for subject in pd_fpower.subject.unique():
    for session in pd_fpower.loc[pd_fpower.subject == subject].session.unique():
        for run in pd_fpower.loc[(pd_fpower.subject == subject)&(pd_fpower.session==session)].run.unique():
            d = pd_fpower.loc[(pd_fpower.subject == subject)&(pd_fpower.session==session)&(pd_fpower.run == run)].set_index(['subject', 'session', 'run', 'segment'])
            l = d.length.values
            d = d.drop('length', axis=1)
            balanced_avg = (d.values.T * l).T.mean(axis=0)/(l.sum()/2)
            d = dict(zip(d.columns, balanced_avg))
            d['subject'] = subject
            d['session'] = session
            d['run'] = run
            d['length'] = l.sum()
            di.append(d)
pd_fpower_avgrun = pd.DataFrame(di)


di = []
for subject in pd_fpower_avgrun.subject.unique():
    for session in pd_fpower_avgrun.loc[pd_fpower_avgrun.subject == subject].session.unique():
        d = pd_fpower_avgrun.loc[(pd_fpower_avgrun.subject == subject)&(pd_fpower_avgrun.session==session)].set_index(['subject', 'session', 'run'])
        l = d.length.values
        d = d.drop('length', axis=1)
        balanced_avg = (d.values.T * l).T.mean(axis=0)/(l.sum()/2)
        d = dict(zip(d.columns, balanced_avg))
        d['subject'] = subject
        d['session'] = session
        d['length'] = l.sum()
        di.append(d)
pd_fpower_avg = pd.DataFrame(di)
```

### Hippus Peak

```python
hippus=[]
hip_limit = .9  
```

```python
f, ax = plt.subplots(5, 4, figsize=(10, 8))
f.tight_layout()
data=psp_baseline.drop(['session'], axis=1)

i=0
for alist in ax:
    for a in alist:
        try:
            sub = data.index[i]
            d = data.loc[sub]
            d = d.loc[(d.index >= 1/32)&(d.index <= 1)]
            X = d.index.astype(float)
            y = d.values

            popt, pcov = curve_fit(func_powerlaw, X, y, p0 = np.asarray([-1,10**5,0]), maxfev=2000)

            a.plot(X, func_powerlaw(X, *popt), '--', color='green')

            d1 = d.loc[(d.index > 1/10)&(d.index < 1)] # q&d way to exclude maxima in the descending power law spectrum
            X1 = d1.index.astype(float)
            y1 = d1.values
            hippus_freq = X1[np.argmax(y1-func_powerlaw(X1, *popt))]
            power_offset = np.max(y1-func_powerlaw(X1, *popt))
            a.axvline(hippus_freq, color='black')
            di = {'subject':sub,
                          'group':'baseline',
                          'frequency':hippus_freq,
                          'offfset':power_offset}
            if hippus_freq > hip_limit:
                di['frequency'] = np.nan
            hippus.append(di)
            a.set(title='Subject {}'.format(sub))


            a.plot(d.index, d.values, color='grey')
            i=i+1
        except IndexError:
            continue
```

```python
f, ax = plt.subplots(6, 4, figsize=(12, 10))
f.tight_layout()
data=ctrl_fpower_avg.drop('session', axis=1)

i=0
for alist in ax:
    for a in alist:
        try:

            sub = data.index[i]
            d = data.loc[sub]

            d = d.loc[(d.index > 1/32)&(d.index < 1)]
            X = d.index.astype(float)
            y = d.values

            popt, pcov = curve_fit(func_powerlaw, X, y, p0 = np.asarray([-1,10**5,0]), maxfev=2000)

            a.plot(X, func_powerlaw(X, *popt), '--', color='green')
            d1 = d.loc[(d.index > 1/10)&(d.index < 1)] # q&d way to exclude maxima in the descending power law spectrum
            X1 = d1.index.astype(float)
            y1 = d1.values
            hippus_freq = X1[np.argmax(y1-func_powerlaw(X1, *popt))]
            power_offset = np.max(y1-func_powerlaw(X1, *popt))
            power = y1[np.argmax(y1-func_powerlaw(X1, *popt))]
            a.axvline(hippus_freq, color='black')
            di = {'subject':sub,
                          'group':'control',
                          'frequency':hippus_freq,
                          'offfset':power_offset,
                 'hippus_power': power}
            if hippus_freq > hip_limit:
                di['frequency'] = np.nan
            hippus.append(di)
            a.set(title='Subject {}'.format(sub))

            a.plot(d.index, d.values, color='grey')
            i=i+1
        except IndexError:
            continue
```
